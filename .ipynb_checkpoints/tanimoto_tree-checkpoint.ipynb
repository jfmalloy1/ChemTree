```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
from skbio import DistanceMatrix
from skbio.tree import nj
```

# Heat Map

Build a heat map of tanimoto similarities (based on default RDKit fingerprints)

```python
""" Creates a heatmap given a filename of a pickled np matrix """
def heatmap(fp):
    #Read in matrix
    m = pickle.load(open(fp, "rb"))
    
    #Make it symmetrical (for visualizations)
    i_lower = np.tril_indices(len(m), -1)
    m[i_lower] = m.T[i_lower]
    
    plt.imshow(m)
```

```python
heatmap("tanimoto_matrix_Orgacids.p")
```

```python
heatmap("tanimoto_matrix_OrgacidsLipids.p")
```

```python
heatmap("tanimoto_matrix_OrgacidsAntibiotics.p")
```

```python
heatmap("tanimoto_matrix_LipidsAntibiotics.p")
```

# Trees

Use the neighbor-joining algorithm (given Tanimoto similarity) to build chemical phylogenetic trees

```python
fp = "tanimoto_matrix_OrgacidsAntibiotics.p"

#Read in matrix
m = pickle.load(open(fp, "rb"))
#Make it symmetrical (for visualizations)
i_lower = np.tril_indices(len(m), -1)
m[i_lower] = m.T[i_lower]

#Convert m to an integer matrix (needed for neighborjoin algorithm, apparently)
m=m*1000
m = m.astype(int)
print(m)
```

```python
#Set up distance matrix
ids = np.arange(0,len(m),1)
dm = DistanceMatrix(m,ids)
print(dm)
print(type(dm))
```

```python
tree = nj(dm)
```
